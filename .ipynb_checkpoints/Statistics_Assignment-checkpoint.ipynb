# Q9: Summarize the differences between the Z-test and the T-test in three sentences.

- The Z-test utilizes the Z-distribution or standard normal distribution, whereas the T-test utilizes the T-distribution.
- The T-test calculates T-statistics based on the degrees of freedom, whereas the Z-test does not require the determination of degrees of freedom.
- When dealing with independent samples that have equal variance, it is recommended to use t-statistics instead of z-tests. This is because the z-test is only applicable when the populations do not differ significantly in their standard deviations.

resource : https://vitalflux.com/when-to-use-z-test-vs-t-test-differences-examples/

## Calculate

Let's assume that a class's average score is higher than 70 with a standard deviation of 10.
Calculate the Z-value to determine whether there is sufficient data to support this claim at a 0.05 
significance level if a sample of 50 students was chosen and their mean score was 80.

```python
p_Mean = 70 #population mean
sd=10 # standard deviation
s=50 # students
s_Mean =80 #sample mean
sign_level= 0.05 

critical_value = stats.norm.ppf(1 - sign_level)
z_value = (s_Mean - p_Mean) / (sd / math.sqrt(s))

print('The Z-value ',z_value )
print("Critical value:", critical_value,"\n")



if z_value > critical_value:
    print("According to the given data, it has been proven that there is a sufficient data to support this claim" )
else:
    print("According to the given data, it has been proven that there is NO sufficient data to support this claim")

```

An establishment wants to boost sales. According to past sales data, each sale made by 30 salesmen averaged 
40 dollar. The most recent data indicated that an average sale per transaction was 60 dollar after some training.Find the t-value for a 20 dollar standard deviation. Did a training program increase sales?

```python
p_Mean = 40 #population mean
sd=20 # standard deviation
s=30 # students
s_Mean =60 #sample mean

critical_value = stats.t.ppf(1 - sign_level/2, (s- 1))
t_value = (s_Mean - p_Mean) / (sd / math.sqrt(s))

print('The t-value =',t_value )
print("Critical value:", critical_value,"\n")

if z_value > critical_value:
    print("According to the given data, it has been proven that there is a sufficient data to support this claim" )
else:
    print("According to the given data, it has been proven that there is NO sufficient data to support this claim")
```
